```python
# Render our plots inline
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# Make the graphs a bit prettier, and bigger
pd.options.display.mpl_style = 'default'
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
import requests
import json
from pandas.tseries.tools import to_datetime

url = "https://crowd-power.appspot.com/_ah/api/mturk/v1/arrivalCompletions/list?from=06%2F01%2F2014&to=02%2F28%2F2015"

resp = requests.get(url)
results = json.loads(resp.text)
df = pd.DataFrame(results["items"])


# Create an index using the date, and drop the from/to columns
df['date'] = map(to_datetime, df['from'])
df.set_index(['date'], inplace=True)
df.drop(["from"],inplace=True,axis=1)
df.drop(["to"],inplace=True,axis=1)

# Drop unecessary columns
df.drop(["id"],inplace=True,axis=1)
df.drop(["kind"],inplace=True,axis=1)

# Convert data types to floats
df['hitsArrived'] = df['hitsArrived'].astype(float)
df['hitsCompleted'] = df['hitsCompleted'].astype(float)
df['hitsAvailableUI'] = df['hitsAvailableUI'].astype(float)

df['hitGroupsArrived'] = df['hitGroupsArrived'].astype(float)
df['hitGroupsCompleted'] = df['hitGroupsCompleted'].astype(float)
df['hitGroupsAvailableUI'] = df['hitGroupsAvailableUI'].astype(float)

# Convert to float, and convert cent values to dollars
df["rewardsCompleted"] = df['rewardsCompleted'].astype(float)
df["rewardsCompleted"] = map(lambda x: x/100, df["rewardsCompleted"])
df["rewardsArrived"] = df['rewardsArrived'].astype(float)
df["rewardsArrived"] = map(lambda x: x/100, df["rewardsArrived"])

# Create a few more columns 

def percWorkCompleted(hitsCompleted, hitsAvailableUI):
    return hitsCompleted/(hitsAvailableUI)

def percWorkPosted(hitsArrived, hitsAvailableUI):
    return hitsArrived/(hitsAvailableUI)

def diffRewards(rewardsCompleted, rewardsArrived):
    return rewardsArrived-rewardsCompleted

df["percHitsCompleted"] = map(percWorkCompleted, df["hitsCompleted"], df["hitsAvailableUI"])
df["percHitsPosted"] = map(percWorkPosted, df['hitsArrived'], df["hitsAvailableUI"])

df["diffHits"] = map(lambda x,y: x-y, df['hitsArrived'], df["hitsCompleted"])
df["hitsAvailableDerived"] = df["diffHits"].cumsum() + 200000
df["diffHitsUI"] = df["hitsAvailableUI"].diff()


df["diffGroups"] = map(lambda x,y: x-y, df['hitGroupsArrived'], df["hitGroupsCompleted"])
df["diffGroupsUI"] = df["hitGroupsAvailableUI"].diff()

df["diffRewards"] = map(lambda x,y: x-y, df['rewardsArrived'], df["rewardsCompleted"])
```

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(df[["hitsAvailableUI", "hitsAvailableDerived" ]], alpha=0.1, figsize=(10, 10), diagonal='kde')
```

```python
# The results indicate that as more HITs arrive, a *higher* *percentage* of the available work
# gets done. This clearly indicates that the arrival of new work also attracts new workers
# in the market, who also seem to spillover to other tasks
import statsmodels.formula.api as sm
result = sm.ols(formula="hitsAvailableDerived ~ hitsAvailableUI", data=df).fit()

print result.params
```

```python
# It seems that there is significant memory for the market, as the autocorrelation of the 
# HITS available (as reported by Amazon) lasts for approximately 7-10 days


from pandas.tools.plotting import autocorrelation_plot

autocorrelation_plot(df["hitsAvailableUI"])
```

```python
# Rolling weekly work volume (weekly, to remove as much as possible the periodicity and memory)

hour_window = 168
coef = 3 # how many stdev to plot

c = pd.Series(data = df["hitsAvailableUI"], index=df.index)
mac = pd.rolling_mean(c, hour_window)
mstdc = pd.rolling_std(c, hour_window)

plt.figure()

plt.plot(df.index, df["hitsAvailableUI"], 'b', alpha=0.2)
plt.plot(mac.index, mac, 'b', alpha=0.9)
plt.fill_between(mstdc.index, mac-coef*mstdc, mac+coef*mstdc, color='b', alpha=0.05)
```
