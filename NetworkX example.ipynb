```python
import networkx as nx
```

```python
def readGraph(file="data/graph_subset.txt"):
    
    
    G=nx.complete_graph(736)
    
    datafile = open(file,"r")
    lines = (line.rstrip('\r\n') for line in datafile)
    n=0
    for line in lines:
        arr=line.split(' ')
        
        # Notice that I convert the node names to integers, for convenience
        # You can keep them as strings as well, but it is awkward to have
        # numbers are nodenames and keep them represented as strings
        white=int(arr[1])
        black=int(arr[0])
        
        G.add_edge(white, black)
            
    return G
```

```python
G=readGraph()
```

```python
ngames=G.number_of_edges()
nplayers=G.number_of_nodes()

print("Loaded %d chess games between %d players\n"          % (ngames,nplayers))
```

```python
# We are testing the adamic_adar index for the pair of nodes: (16393,16399)
# We can test other link prediction functions as well: 
# See https://networkx.github.io/documentation/latest/reference/algorithms.link_prediction.html

score = nx.adamic_adar_index(G,[(16393,16399)])
for u, v, p in score:
    print '(%d, %d) -> %.8f' % (u, v, p) 
```

```python
# We now compute the similarity scores for all node pairs, and print only the node pairs
# with score above 0.9
score = nx.adamic_adar_index(G)
for u,v,p in score:
    if (p>0.9):
        print '(%d, %d) -> %.8f' % (u, v, p)
```
